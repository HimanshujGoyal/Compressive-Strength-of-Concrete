```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,SCORERS
from sklearn.model_selection import cross_val_score,cross_val_predict,KFold,train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
sorted(SCORERS.keys())
```

```python
def cross_val(model):
    pred = cross_val_score(model,X_Scaled,y_concat,cv=10,scoring='neg_mean_squared_error')
    return pred.mean()

def print_scores(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mse)
    r_squared = r2_score(true,predicted)
    print(f'MAE is {mae}')
    print(f'MSE is {mse}')
    print(f'RMSE is {rmse}')
    print(f'R_squared is {r_squared}')
    print('----------------------------')
    
def evaluate(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mse)
    r_squared = r2_score(true,predicted)
    return mae,mse,rmse,r_squared    
```
